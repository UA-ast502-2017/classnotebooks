````python
import requests
from astropy.io import fits
from os.path import exists, sep
from os import makedirs

def retrieve_fits(url, base_directory='data'):
    makedirs(base_directory, exist_ok=True)
    local_filename = base_directory + sep + url.split('/')[-1]
    if not exists(local_filename):
        r = requests.get(url, stream=True)
        with open(local_filename, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1024): 
                if chunk: # filter out keep-alive new chunks
                    f.write(chunk)
    return fits.open(local_filename)
````

````python
DATA_ARCHIVE_URL = 'https://archive.stsci.edu/missions/hlsp/laplace/dd1/LAPL/NICMOS-LAPL-DD1/LAPL_DATA/contemp_flats/repaired/'
file_listing = requests.get(DATA_ARCHIVE_URL).text
````

````python
import re
````

````python
file_urls = [
    DATA_ARCHIVE_URL + fn
    for fn in re.findall(r'<a +href="([^"]+.fits)', file_listing)
]
````

````python
for fn in file_urls:
    hdul = retrieve_fits(fn, base_directory='data/laplace/dd1/contemp_flats_repaired/')
    print(fn)
````
